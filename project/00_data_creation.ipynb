## Importing Packages

```python
import numpy as np
import pandas as pd
```

## Reading In Full Data Set

```python
df_ml = pd.read_csv("../data/data_ml.csv")
```

```python
df_ml = pd.read_csv('../data/data_ml.csv')
pd.set_option('display.max_rows', 101)
df_ml.head().T
```

## Separating out Training and Testing Set By Date

```python
separation_date = "2013-12-31"
df_train = df_ml.query("date < @separation_date").copy()
df_test = df_ml.query("@separation_date <= date & date < '2018-12-31'").copy()
```

```python
np.sort(df_train["date"].unique())
```

## Dropping Unused Labels from Training Set

```python
training_labels_to_drop = ["R1M_Usd", "R3M_Usd", "R6M_Usd", "R12M_Usd", "R12M_Usd_C"]
df_train.drop(columns=training_labels_to_drop, inplace=True)
```

```python
df_train.head().T
```

## Changing the Ordering of the Training Set

This is just to make sure that the students don't use returns from the in-class data set.

```python
df_train = df_train.sample(n=len(df_train), random_state=1234)
```

## Dropping Unused Labels from Testing Set

```python
testing_labels_to_drop = ["R3M_Usd", "R6M_Usd", "R12M_Usd", "R1M_Usd_C","R12M_Usd_C"]
df_test.drop(columns=testing_labels_to_drop, inplace=True)
```

```python
df_test.head().T
```

## Writing Training and Testing Set to CSV

```python
df_train.to_csv("mlfactor_project_training.csv", index=False)
```

```python
df_test.to_csv("mlfactor_project_testing.csv", index=False)
```
