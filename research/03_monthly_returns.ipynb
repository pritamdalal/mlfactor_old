```python
import numpy as np
import pandas as pd
```

```python
df_ml = pd.read_csv("../data/data_ml.csv")
```

```python
separation_date = "2013-12-31"
df_train = df_ml.query("date < @separation_date")
df_test = df_ml.query("@separation_date <= date & date < '2018-12-31'")
```

```python
np.sort(df_train['date'].unique())
```

```python
np.sort(df_test['date'].unique())
```

```python
df_monthly_returns = df_test.groupby(['date'])[['R1M_Usd']].mean().reset_index()
df_monthly_returns
```

```python
df_monthly_returns['growth_factor'] = 1 + df_monthly_returns['R1M_Usd']
df_monthly_returns
```

```python
df_monthly_returns['equity_curve'] = df_monthly_returns['growth_factor'].cumprod()
df_monthly_returns
```

```python
df_monthly_returns.plot(x='date', y='equity_curve', figsize=(12, 4));
```
